```python
title_data3abc = []
question_data3abc = []
user_name3abc = []
user_type3abc = []
date_data3abc = []
link_url3abc = []

# driver = webdriver.Chrome('chromedriver.exe')
chrome_options = Options()
chrome_options.add_argument('disable-notifications')
chrome_options.add_argument("enable-automation")
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")
chrome_options.add_argument("--disable-browser-side-navigation")
chrome_options.add_argument('disable-infobars')
chrome_options.add_argument("--disable-gpu")
chrome_options.add_argument("--lang=en-US")

actions = ActionChains(driver)
driver = webdriver.Chrome(service=Service('chromedriver.exe'), options=chrome_options)
 
url = "https://www.alodokter.com/komunitas/diskusi/penyakit/page/4000"

driver.get(url)
# driver.maximize_window()

# Loop for every page we want
for i in range(0, 64):
    # We check how many topics we have in this page

    how_many_topics = WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.XPATH, '//*[@id="topic-list"]/card-topic')))

    #Loop for every topic
    for i in range(1, len(how_many_topics)):
        # We click on the topic link
        time.sleep(0.5)
        topic = driver.find_element(By.XPATH, '//*[@id="topic-list"]/card-topic[{}]'.format(i))
        # topic = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, f'(//*[@id="topic-list"]/card-topic)[{i}]')))
        time.sleep(0.5)
        shadow1 = driver.execute_script("return arguments[0].shadowRoot", topic)
        # click_topic = WebDriverWait(shadow1, 10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, 'a')))
        time.sleep(0.5)
        shadow1.find_element(By.CSS_SELECTOR, 'a').send_keys(u'\ue007')
        
        # shadow1.find_element(By.CSS_SELECTOR, 'a').click() 
        # We get the details we want, in my example only the title, but you can take whatever you want
        #tittle
        
        detail_topic = WebDriverWait(driver, 100).until(EC.presence_of_element_located((By.CSS_SELECTOR, 'detail-topic')))
        shadow2 = driver.execute_script("return arguments[0].shadowRoot", detail_topic)
        title = shadow2.find_element(By.CSS_SELECTOR, '.h2').text
        title_data3abc.append(title)
    
        #question
        qt = shadow2.find_element(By.CSS_SELECTOR, 'user-topic')
        shadow3 = driver.execute_script("return arguments[0].shadowRoot", qt)
        qt_ele = shadow3.find_element(By.CSS_SELECTOR, 'p')
        qt_text = qt_ele.text
        question_data3abc.append(qt_text)

        #getting user_detail
        user = driver.execute_script('return arguments[0].shadowRoot', qt)
        user_ele = user.find_element(By.CLASS_NAME, 'user-topic-title')

        #user_name
        user_name_ele = user_ele.find_element(By.TAG_NAME, 'h4')
        user_name_text = user_name_ele.text
        user_name3abc.append(user_name_text)

        #user_type
        user_type_el = user_ele.find_element(By.TAG_NAME, 'span')
        user_type_text = user_type_el.text
        user_type3abc.append(user_type_text)
        
        #date
        date = driver.execute_script('return arguments[0].shadowRoot', qt)
        date_ele = date.find_element(By.CLASS_NAME, 'user-topic-count')

        date_ymd = date_ele.find_element(By.CLASS_NAME, 'small-text')
        date_text = date_ymd.text.strip()
        date_data3abc.append(date_text)

        #get_url
        link = driver.current_url
        link_url3abc.append(link)

        # We click on back button
        time.sleep(0.5)
        driver.execute_script("window.history.go(-1)")
    # We search for the next button and we click on it
    pagination = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, f'//paginate-button')))
    shadow_pagination = driver.execute_script("return arguments[0].shadowRoot", pagination)
    next_button = shadow_pagination.find_element(By.CSS_SELECTOR, 'a.page-next')
    actions.move_to_element(next_button)
    time.sleep(0.5)
    actions.click(next_button).perform()
     
    
```
